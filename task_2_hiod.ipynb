```python
import pandas as pd
```

```python
link = 'https://storage.yandexcloud.net/lms-itmo-ru-files-27a87tyf/Data_storage_and_processing/Ind_2/task2_363981.csv'
df = pd.read_csv(link)
df
```

```python
df['<PRICE>'] = (df['<OPEN>'] + df['<CLOSE>'] + df['<LOW>'] + df['<HIGH>'])/4
```

```python
filtered_row = df[(df['<DATE>'] == '03.11.2020') & (df['<TIME>'] == '11:30:00')]
filtered_row
```

```python
df['<TOTAL>'] = df['<PRICE>']*df['<VOL>']

filtered_rows = df[df['<DATE>'] == '30.11.2020']

# Вычисляем суммарный оборот (предположим, что оборот хранится в столбце '<VOL>')
total_turnover = filtered_rows['<TOTAL>'].sum()
t1 = filtered_rows.sum()
print(total_turnover, t1)
```
