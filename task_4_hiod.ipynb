```python
import pandas as pd
import numpy as np
import re
```

```python
link = 'https://storage.yandexcloud.net/lms-itmo-ru-files-27a87tyf/Data_storage_and_processing/Ind_4_2/task4_1_873933.csv'
df = pd.read_csv(link, index_col = 'ID', sep = ';')
df
```

for Artem

```python
data = df[['DISTANCE', 'STOP_COUNT', 'COST', 'FITNESS', 'INTERNET']]
data = data[data['INTERNET']!= 0]
data = 1 - np.exp(1-data/data.min())
data['SUM']=data['STOP_COUNT'] + data['DISTANCE'] + data['COST'] + 0.3*data['FITNESS'] - 0.3*data['INTERNET']
data.sort_values(by = ['SUM'])
```

for ANNA

```python

data = df[['DISTANCE', 'STOP_COUNT', 'COST', 'INTERNET', 'PETS_ALLOWED', 'ADDITIONAL_INFO']]
data = data[data['INTERNET'] != 0]
data = data[data['PETS_ALLOWED'] == 'Да']
data
```

```python
data['ADDITIONAL_INFO'] = data['ADDITIONAL_INFO'].fillna('')
pattern = re.compile(r'(площадк|выгул)(.*?)(питом|собак|четвероног)', re.IGNORECASE)
data['DOG_WALKING'] = data['ADDITIONAL_INFO'].apply(lambda x: 1 if re.search(pattern, x) else 0)
data
```

```python
data = data.drop(['PETS_ALLOWED', 'ADDITIONAL_INFO'], axis = 1)
data
```

```python
data = (data - data.min()) / (data.max() - data.min())
data
```

```python
data['SUM'] = data['STOP_COUNT'] + data['DISTANCE'] + data['COST'] -0.2*data['INTERNET'] -0.2*data['DOG_WALKING']
data.sort_values(by = 'SUM')
```
