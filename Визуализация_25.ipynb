```python
import folium
from folium.plugins import HeatMap
```

```python
from geopy.geocoders import Nominatim
```

```python
loc = Nominatim(user_agent="GetLoc")
getLoc = loc.geocode("Ankara")
print(getLoc.address)
coordinates = [getLoc.latitude, getLoc.longitude]
print(coordinates)
```

```python
data1 = list(map(list, zip(df_Turkey['latitude'],df_Turkey['longitude'],df_Turkey['mag'])))
data1[:5]
```

```python
map_obj = folium.Map(location = coordinates,zoom_start = 5)
#,tiles='StamenTerrain')
HeatMap(data1).add_to(map_obj)
map_obj
```

```python
from ChernoffFace import *
```

```python
df_all = pd.read_excel('University.xlsx')
df_all
```

```python
df_all[['Качество образования, %',"Востребованность выпускников, %",'Качество нетворкинга, %','Международное признание, %']] = df_all[['Качество образования, %',"Востребованность выпускников, %",'Качество нетворкинга, %','Международное признание, %']]/100
df_all
```

```python
fig = chernoff_face(data=df_all,
                    n_columns=5,
                    long_face=False,
                    figsize=(6, 1), dpi=200)
#color_mapper=plt.cm.rainbow,
#plt.cm.Paired
plt.show()
```
